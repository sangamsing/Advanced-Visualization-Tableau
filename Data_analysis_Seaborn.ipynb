## Data Visualiation


###Univariate -Analysis
Continuous variable

```python
data['tip']
```

```python
sns.distplot(data['tip'])
```

```python
sns.distplot(data['tip'],hist=False)
```

```python
sns.distplot(data['tip'],hist=False,axlabel="Tip given by Customer")
```

```python
sns.distplot(data['tip'],hist=False,axlabel="Tip given by Customer",color="red")
```

```python
sns.histplot(data['tip'])
```

```python
sns.boxplot(y=data['tip'])
```

###univariate Analysis

```python
sns.countplot(data["smoker"])
```

###Bi-variate Analysis
1. both are contiuous
2. both categorical
3. one is contiuous and other is categorical


```python
#1.both are contiuous
sns.scatterplot(data=data,x='tip',y="total_bill")
```

```python
sns.lineplot(data=data,x='tip',y="total_bill")
```

```python
#2. both are categorical
sns.countplot(data=data,x="sex",hue="time")
```

```python
#3. one is continuous other is categorical
#boxplot
sns.boxplot(data=data,y="tip",x="time")
```

```python
sns.histplot(data=data,x="tip",hue="time")
```

```python
#Mulitvariate
#1.heatmap
sns.heatmap(data.corr(),annot=True)
```

```python
sns.pairplot(data)
```

```python
#use countplot instead
sns.catplot(data=data,x="sex",y="time",hue="smoker")
```
